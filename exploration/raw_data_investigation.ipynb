```python
from collections import defaultdict
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
```

```python
root_data_dir = Path('../dataset')
assert root_data_dir.exists()
```

```python
celeb_num_images = defaultdict(lambda: {'train': 0, 'test': 0})
# for train data
for celeb_dir in (root_data_dir / 'train').iterdir():
    celeb_id = int(celeb_dir.name)
    num_img = len(list(celeb_dir.iterdir()))
    celeb_num_images[celeb_id]['train'] = num_img

# for test data
for celeb_dir in (root_data_dir / 'test').iterdir():
    celeb_id = int(celeb_dir.name)
    num_img = len(list(celeb_dir.iterdir()))
    celeb_num_images[celeb_id]['test'] = num_img
```

```python
celeb_num_images_df = pd.DataFrame(
    {'train': [v['train'] for v in celeb_num_images.values()],
     'test': [v['test'] for v in celeb_num_images.values()]},
     index=celeb_num_images.keys())
```

```python
sns.histplot(celeb_num_images_df, x='train')
plt.title('Number of images per celebrity in train data')
```

```python
sns.histplot(celeb_num_images_df, x='test')
plt.title('Number of images per celebrity in test data')
```
